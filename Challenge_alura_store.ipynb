#1. Análise do faturamento


```python
locais = loja["Local da compra"].iloc[0:5]
print(locais)
```

```python
faturamento_total =(loja["Preço"].iloc[0:5]*loja["Quantidade de parcelas"].iloc[0:5])
faturamento_total
```

```python
plt.bar(x = locais, height = faturamento_total )
plt.title("Faturamento por Local")
plt.xlabel("Local")
plt.ylabel("Faturamento")
plt.show()
```
